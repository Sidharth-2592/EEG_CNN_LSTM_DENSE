```python
# Importing libraries 

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import Dropout
from tensorflow import keras
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from keras.layers import Flatten
from keras.layers import TimeDistributed
from keras.layers import BatchNormalization
from tensorflow.keras.utils import  plot_model
from keras.layers import Input
from keras import Model
```

```python
# Importing EEG Features and Kinematic Data
from google.colab import drive
drive.mount('/content/drive')
EEG_features = pd.read_csv('/content/drive/MyDrive/Trajectory Estimation CSV Files/EEG_features.csv') 
kin_Y = pd.read_csv('/content/drive/MyDrive/Trajectory Estimation CSV Files/kin_Y.csv') # x coordinates: Kin_X; y coordinates: Kin_Y; z coordinates: Kin_Z;
```

```python
print(EEG_features.shape)
print(kin_Y.shape)
```

```python
from sklearn.model_selection import train_test_split

x=EEG_features.iloc[0:29000,:352].values
y=kin_Y.iloc[0:29000,-1].values

trainX, testX, trainy, testy = train_test_split(x, y, test_size = 0.2) # Spliting features: Training 80% and Testing 20%

Cross_x=EEG_features.iloc[29001:-1,:352] # Keeping few trails for cross checking model's performance
cross_y=kin_Y.iloc[29001:-1,-1].values

trainX1 = trainX.reshape((trainX.shape[0], 1,352, 1))
testX1= testX.reshape((testX.shape[0], 1, 352, 1))
```

```python
# Defining model: Time distributed CNN LSTM model

def CNN_LSTM_Model(x):
  x=TimeDistributed(Conv1D(filters=64, kernel_size=14, activation='relu',
                                  input_shape=(None, 352,1)))(x)
  x=TimeDistributed(Dropout(0.1))(x)
  x=TimeDistributed(BatchNormalization())(x)
  x=TimeDistributed(Conv1D(filters=32, kernel_size=7, activation='relu'))(x)
  x=TimeDistributed(MaxPooling1D(pool_size=2))(x)
  x=TimeDistributed(Flatten())(x)
  x=LSTM(50, activation='relu')(x)
  x=Dropout(0.1)(x)
  x=BatchNormalization()(x)
  x=Dense(50)(x)
  x=Dropout(0.1)(x)
  x=BatchNormalization()(x)
  output=Dense(1)(x)
  return output

# The value of epoch can be 30,20,10,5 or 50,50,50,50 in line 211,212,213,214 respectively
```

```python
input =Input(shape=(None, 352,1))
output=CNN_LSTM_Model(input)
```

```python
# Ploting model

model= Model(input, output)
plot_model(model)
```

```python
# Pearson correlction metric 

from keras import backend as K
def pearson_r(y_true, y_pred):
    x = y_true
    y = y_pred
    mx = K.mean(x, axis=0)
    my = K.mean(y, axis=0)
    xm, ym = x - mx, y - my
    r_num = K.sum(xm * ym)
    x_square_sum = K.sum(xm * xm)
    y_square_sum = K.sum(ym * ym)
    r_den = K.sqrt(x_square_sum * y_square_sum)
    r = r_num / r_den
    return K.mean(r)
```

```python
model.compile(optimizer='Adam',  loss='mse',  metrics=['mse', pearson_r])
```

```python
model.fit(trainX1, trainy, batch_size=50,validation_data=(testX1, testy), epochs=10, verbose=2)
model.fit(trainX1, trainy, batch_size=100,validation_data=(testX1, testy), epochs=10, verbose=2)
model.fit(trainX1, trainy, batch_size=200,validation_data=(testX1, testy), epochs=10, verbose=2)
model.fit(trainX1, trainy, batch_size=400,validation_data=(testX1, testy), epochs=10, verbose=2)
```

```python
x_input = np.array(Cross_x)
x_input = x_input.reshape((x_input.shape[0], 1, 352, 1))
yhat = model.predict(x_input, verbose=1)
plt.figure(figsize=(30,5))
plt.rcParams['font.size'] = '20'
plt.xticks(rotation = 90)
plt.xlabel('Samples')
plt.plot(cross_y)
plt.plot(yhat)
plt.show()
```
